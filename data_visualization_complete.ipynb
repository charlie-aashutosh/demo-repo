```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
#loading the iris dataset from sklearn liabrary
iris=snr.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.tail(10)
```

```python
#line graph
ax=iris.plot(figsize=(15,8),title='Iris dataset')
ax.set_xlabel('x axis') #total number of rows in the iris dataset
ax.set_ylabel('y axis') #highest magnitude of numerical variables 
```

```python
#drop the species column 
data=iris.drop(['species'],axis=1) #axis=1 means entire column will be deleted from the dataset
```

```python
#tips dataset
new_data=snr.load_dataset('tips')
```

```python
snr.relplot(x='total_bill',y='tip',data=new_data,hue='smoker',style='time')
```

```python
my_data=pd.read_csv("/content/Advertising.csv")
```

```python
data_new=my_data.drop(['Unnamed: 0'],axis=1)
```

```python
data_new.head()
```

```python
data_new.plot(kind='bar')
```

```python
data_new.iloc[0:2].plot(kind='bar')
```

```python
data_new.iloc[0:4].plot(kind='bar')
```

```python
data_new.iloc[0:2].plot.pie(subplots=True,figsize=(10,10),fontsize=10,autopct='%.2f')
```

```python
#pairplot
snr.pairplot(data_new,x_vars=['TV','Radio','Newspaper'],y_vars=['Sales'],size=10)
```

```python
#heatmap - will show the corelation between features
plt.figure(figsize=(5,5)) 
snr.heatmap(data_new.corr(),annot=True)
```
